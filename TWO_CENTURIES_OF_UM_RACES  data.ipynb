```python
#what age groups are the best in the 50m race (20 + races win)
```

```python
df3.query('race_length  == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False).query('count>19').head()
```

```python
#what age groups are the worst in the 50m race (10 + races win)
```

```python
df3.query('race_length  == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = True).query('count>9')
```

```python
#seasons for the data -> Slower in summer than winter?



#Spring
#Summer
#fall
#Winter

#split between two decimals

```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x:  'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```
